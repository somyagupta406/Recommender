```python
import pandas as pd
import numpy as np
```

```python
M = pd.read_csv("Movie Reviews - Sheet1.csv", index_col='Name')
```

```python
M
```

```python
def pearson(s1, s2):
    """Take two pd.Series objects and return a pearson correlation."""
    s1_c = s1 - s1.mean()
    s2_c = s2 - s2.mean()
    return np.sum(s1_c * s2_c) / np.sqrt(np.sum(s1_c ** 2) * np.sum(s2_c ** 2))
```

```python
def get_recs(movie_name, M, num):

    import numpy as np
    reviews = []
    for title in M.columns:
        if title == movie_name:
            continue
        cor = pearson(M[movie_name], M[title])
        if np.isnan(cor):
            continue
        else:
            reviews.append((title, cor))
    
    reviews.sort(key=lambda tup: tup[1], reverse=True)
    return reviews[:num]
```

### Question 1: What movie is most Similar to 'The fault in our stars' (60 pts)

```python
recs = get_recs('The Fault in Our Stars', M, 5)
```

```python
#This will predict movie rating by applying mean
def get_ratings(movie):
    predictedMovies=get_recs(movie,M,1)
    movie_title=[i[0] for i in predictedMovies]
    rating_movie = M[movie_title].mean()
    rating=(movie,round(rating_movie))
    return rating
```
